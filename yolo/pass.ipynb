```python
from pathlib import Path
import cv2

cropped_dir = Path("dataset/cropped_rgb")
mask_dir    = Path("dataset/masks")

mismatches = []
for img_path in sorted(cropped_dir.glob("*.png")):
    mask_path = mask_dir / img_path.name
    if not mask_path.exists():
        print(f"[MISSING MASK] {img_path.name}")
        continue
    img  = cv2.imread(str(img_path))
    mask = cv2.imread(str(mask_path), cv2.IMREAD_GRAYSCALE)
    if img.shape[:2] != mask.shape[:2]:
        mismatches.append((img_path.name, img.shape[:2], mask.shape[:2]))

if mismatches:
    print(f"\n❌ {len(mismatches)} mismatches found:")
    for name, i, m in mismatches:
        print(f"  {name}: image={i}  mask={m}")
else:
    print(f"✅ All {len(list(cropped_dir.glob('*.png')))} images match their masks.")
```
